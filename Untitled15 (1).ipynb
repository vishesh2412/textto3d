```python

!pip install torch torchvision torchaudio --quiet
!pip install transformers diffusers trimesh matplotlib open3d numpy scipy plotly --quiet

try:
    import torch
    if torch.cuda.is_available():
        cuda_version = torch.version.cuda.replace(".", "")[:2]
        !pip install -U xformers --index-url https://download.pytorch.org/whl/cu{cuda_version} --quiet
except:
    print("xformers installation skipped - will use default attention")

import numpy as np
import torch
from transformers import CLIPTextModel, CLIPTokenizer
from diffusers import StableDiffusionPipeline, DDIMScheduler
import trimesh
import open3d as o3d
from scipy.ndimage import gaussian_filter
import matplotlib.pyplot as plt
from PIL import Image
import os
from sklearn.neighbors import NearestNeighbors
import plotly.graph_objects as go

OBJECT_TYPE = "chair"
NUM_VIEWS = 6
SEED = 42
MODEL_ID = "stabilityai/stable-diffusion-2-1"

def initialize_pipeline():
    """Initialize the diffusion pipeline with error handling"""
    try:
        scheduler = DDIMScheduler.from_pretrained(MODEL_ID, subfolder="scheduler")
        pipe = StableDiffusionPipeline.from_pretrained(
            MODEL_ID,
            torch_dtype=torch.float16 if torch.cuda.is_available() else torch.float32,
            scheduler=scheduler,
            safety_checker=None,
        ).to("cuda" if torch.cuda.is_available() else "cpu")

        try:
            if torch.cuda.is_available():
                pipe.enable_xformers_memory_efficient_attention()
                print("xformers enabled for memory efficiency")
        except:
            print("xformers not available - using default attention")

        return pipe
    except Exception as e:
        print(f"Pipeline initialization failed: {e}")
        raise

pipe = initialize_pipeline()
generator = torch.Generator(device="cuda").manual_seed(SEED) if torch.cuda.is_available() else None

def generate_views(prompt, num_views):
    """Generate consistent multi-view images"""
    view_angles = [
        ("front", 0, 0),
        ("side", 0, 90),
        ("back", 0, 180),
        ("top", 90, 0),
        ("45-degree", 30, 45),
        ("bottom", -30, 0)
    ][:num_views]

    base_prompt = f"CAD model of {prompt}, white background, orthographic projection, clean edges"
    negative_prompt = "blurry, cropped, lowres, text, watermark, multiple objects"

    images = []
    for view_name, elev, azim in view_angles:
        view_prompt = f"{base_prompt}, {view_name} view"
        image = pipe(
            view_prompt,
            negative_prompt=negative_prompt,
            num_inference_steps=50,
            guidance_scale=7.5,
            height=512,
            width=512,
            generator=generator
        ).images[0]
        images.append((image, elev, azim))

    return images

def create_depth_maps(images):
    """Create normalized depth maps from images"""
    depth_maps = []
    for img, _, _ in images:
        gray = np.array(img.convert('L')) / 255.0
        depth = 1.0 - gray
        depth = gaussian_filter(depth, sigma=3)
        depth_maps.append(depth)
    return depth_maps

def create_point_cloud(depth_maps, images):
    """Create 3D point cloud from multiple views"""
    all_points = []

    for (depth, (_, elev, azim)) in zip(depth_maps, images):
        h, w = depth.shape
        x = np.linspace(-1, 1, w)
        y = np.linspace(-1, 1, h)
        xv, yv = np.meshgrid(x, y)
        z = depth * 2 - 1

        elev_rad = np.deg2rad(elev)
        azim_rad = np.deg2rad(azim)

        x_rot = xv * np.cos(azim_rad) - z * np.sin(azim_rad)
        z_rot = xv * np.sin(azim_rad) + z * np.cos(azim_rad)
        y_rot = yv * np.cos(elev_rad) - z_rot * np.sin(elev_rad)
        z_final = yv * np.sin(elev_rad) + z_rot * np.cos(elev_rad)

        points = np.stack([x_rot, y_rot, z_final], axis=-1)
        all_points.append(points.reshape(-1, 3))

    return np.concatenate(all_points)

def create_mesh(points):
    """Create optimized mesh from point cloud"""
    pcd = o3d.geometry.PointCloud()
    pcd.points = o3d.utility.Vector3dVector(points)

    cl, _ = pcd.remove_statistical_outlier(nb_neighbors=20, std_ratio=2.0)

    cl.estimate_normals()

    radii = [0.01, 0.02, 0.04]
    mesh = o3d.geometry.TriangleMesh.create_from_point_cloud_ball_pivoting(
        cl, o3d.utility.DoubleVector(radii))

    tri_mesh = trimesh.Trimesh(
        vertices=np.asarray(mesh.vertices),
        faces=np.asarray(mesh.triangles)
    )
    tri_mesh = tri_mesh.process(validate=True)
    tri_mesh.remove_degenerate_faces()
    return tri_mesh

def generate_3d_model(prompt, output_base="output"):
    """Complete 3D generation pipeline"""
    print(f"Generating 3D model of: {prompt}")

    print("Creating multi-view images...")
    images = generate_views(prompt, NUM_VIEWS)

    plt.figure(figsize=(15, 8))
    for i, (img, elev, azim) in enumerate(images):
        plt.subplot(2, 3, i+1)
        plt.imshow(img)
        plt.title(f"View {i+1}: Elev={elev}°, Azim={azim}°")
        plt.axis('off')
    plt.tight_layout()
    plt.show()

    print("Processing depth maps...")
    depth_maps = create_depth_maps(images)

    print("Building 3D point cloud...")
    points = create_point_cloud(depth_maps, images)

    print("Generating 3D mesh...")
    mesh = create_mesh(points)

    obj_path = f"{output_base}.obj"
    ply_path = f"{output_base}.ply"
    mesh.export(obj_path)
    mesh.export(ply_path)
    print(f"Saved 3D model to:\n- {obj_path}\n- {ply_path}")

    return mesh

model = generate_3d_model(OBJECT_TYPE, "output_"+OBJECT_TYPE)

def show_3d_model(ply_path):
    mesh = trimesh.load(ply_path)

    fig = go.Figure(data=[
        go.Mesh3d(
            x=mesh.vertices[:,0],
            y=mesh.vertices[:,1],
            z=mesh.vertices[:,2],
            i=mesh.faces[:,0],
            j=mesh.faces[:,1],
            k=mesh.faces[:,2],
            color='lightblue',
            opacity=0.8
        )
    ])

    fig.update_layout(
        scene=dict(
            xaxis=dict(visible=False),
            yaxis=dict(visible=False),
            zaxis=dict(visible=False)
        ),
        width=800,
        height=600
    )
    fig.show()

print("Displaying 3D model...")
show_3d_model(f"output_{OBJECT_TYPE}.ply")

!zip -r 3d_model.zip output_*.*
from google.colab import files
files.download("3d_model.zip")
```
