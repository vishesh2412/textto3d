```python
from diffusers import StableDiffusionPipeline
import torch

# Load the Stable Diffusion model
model = StableDiffusionPipeline.from_pretrained("runwayml/stable-diffusion-v1-5").to("cuda" if torch.cuda.is_available() else "cpu")

# Define angles for different viewpoints
angles = ["front view", "side view", "top view", "back view", "perspective view"]
shoe_description = "a stylish sneaker with modern design, highly detailed, white background"

# Generate images
for angle in angles:
    prompt = f"{shoe_description}, {angle}"
    image = model(prompt).images[0]
    image.save(f"shoe_{angle.replace(' ', '_')}.png")

print("Shoe images generated successfully.")
```
